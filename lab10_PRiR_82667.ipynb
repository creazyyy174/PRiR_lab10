```python
import numpy as np
import time

def f(x):
    return np.sin(x)  # funkcja do całkowania

def rectangle_method(a, b, n):
    h = (b - a) / n  # długość przedziału
    x = np.linspace(a, b, n+1)
    y = f(x)
    return (h * np.sum(y))

a = 0
b = 3.14
n = 10000
start_time = time.time()
result = rectangle_method(a, b, n)
end_time = time.time()
print("Wynik całkowania metodą prostokątów:", result)
print("Czas obliczeń:", end_time - start_time)
```

```python
import numpy as np
import time

def f(x):
    return np.sin(x)  # funkcja do całkowania

def trapezoid_method(a, b, n):
    h = (b - a) / n  # długość przedziału
    x = np.linspace(a, b, n+1)
    y = f(x)
    return (h/2) * (np.sum(y) - y[0]/2 - y[-1]/2)

a = 0
b = 3.14
n = 10000
start_time = time.time()
result = trapezoid_method(a, b, n)
end_time = time.time()
print("Wynik całkowania metodą trapezów:", result)
print("Czas obliczeń:", end_time - start_time)
```

```python
import numpy as np
import time

def f(x):
    return np.sin(x)  # funkcja do całkowania

def simpson_method(a, b, n):
    h = (b - a) / n  # długość przedziału
    x = np.linspace(a, b, n+1)
    y = f(x)
    return (h/3) * (y[0] + 4*np.sum(y[1::2]) + 2*np.sum(y[2:-1:2]) + y[-1])

a = 0
b = 3.14
n = 10000
start_time = time.time()
result = simpson_method(a, b, n)
end_time = time.time()
print("Wynik całkowania metodą Simpsona:", result)
print("Czas obliczeń:", end_time - start_time)
```
